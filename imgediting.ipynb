```python
# Put all pixels for x and y in black
# RGB = 0,0,0
for x in range(my_image.width):
  for y in range(my_image.height):
    my_image.putpixel((x,y), (0,0,0))

my_image
```

```python
# Put all pixels for (x,y) that are even in black
# RGB = 0,0,0
for x in range(my_image.width):
  for y in range(my_image.height):
    if x % 2 == 0 and y % 2 == 0:
      my_image.putpixel((x,y), (0,0,0))

my_image

```

```python
# Pixel average
# TOCOMPLETE
def pixel_average(p):
  return int((p[0]+p[1]+p[2])/3)
```

```python
# Replace bright pixels by grey pixels
# TOCOMPLETE
def filtre_average(img):
    for x in range(img.width):
      for y in range(img.height):
        pixel = img.getpixel((x,y))
        m = pixel_average(pixel)
        if pixel_average(pixel) > 153:
          img.putpixel((x,y), (127, 127, 127)) # Set pixel to grey

filtre_average(my_image)
display(my_image)
```

```python
# Replace R pixels <120 to red pixels
# TOCOMPLETE
for x in range(my_image.width):
  for y in range(my_image.height):
    # Get the RGB value of the current pixel
    current_color = my_image.getpixel((x, y))
    # Check if the pixel is red
    if current_color < (120, 255, 255):
      # Set the pixel to red
      my_image.putpixel((x, y), (255, 0, 0))

my_image
```
